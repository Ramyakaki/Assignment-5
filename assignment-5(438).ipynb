```python
import matplotlib.pyplot as plt

years= ['2013','2014','2015','2016','2017']
values = [10, 15, 8, 12, 18]
plt.plot(years, values)
plt.xlabel('Time')
plt.ylabel('Numeric Variable')
plt.title('Trend of Numeric Variable Over Time')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
correlation_matrix = np.array([[1.0, 0.8, 0.2],
                              [0.8, 1.0, -0.4],
                              [0.2, -0.4, 1.0]])

correlation_matrix 
```

```python
sns.heatmap(correlation_matrix )
```
